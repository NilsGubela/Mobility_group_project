```python
def network_visualization(A, colors, epsilon = 1/333, seed = 11):
    np.random.seed(seed)

    N = len(A)

    col = np.where(colors<1,'k','r')

    xcor = np.arange(0,1,1/N)
    np.random.shuffle(xcor)
    #xcor = [np.random.rand(1)]
    ycor = [np.random.rand(1)]

    epsilon = 1/333

    i = 1
    while i < N:
        print(i)
        #x = np.random.rand(1)
        x = xcor[i]
        y = np.random.rand(1)
        set = False
        for j in range(len(ycor)):
            if (x - xcor[j])**2 + (y - ycor[j])**2 < epsilon:
                set = False 
                break
            else:
                set = True
        if set:
            #xcor.append(x)
            ycor.append(y)
            i +=1 


    fig, ax = plt.subplots()
    ax.scatter(xcor, ycor, s = 10, c = col)
    ax.axis('off')
    
    for i in range(N):
        # only consider upper diagonal
        for j in range(i+1,N):
             if A[i,j] == 1:
                    ax.plot((xcor[i],xcor[j]),(ycor[i],ycor[j]), color = "grey",)
    
    ax.scatter(xcor, ycor, s = 10, c = col)
                
    return(fig, ax)
```
