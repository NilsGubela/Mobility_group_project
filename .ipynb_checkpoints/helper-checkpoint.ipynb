```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.linspace(0,26)
mu = 2
sigma = (26-2)/2 
car = np.exp(((x-mu)/sigma)**2/(-2)) / np.sqrt(2*np.pi * sigma**2)*100
```

```python
plt.plot(x,car)
```

```python
x = np.linspace(0,100,100)
cm = 50
u_max = 4.5

av = u_max/(1+np.exp(-(x-cm))) + 0.5
#av = x/100*u_max
av2 = [np.exp(s/cm) if s < cm else u_max for s in x]
plt.plot(x,av)
plt.plot(x,av2)
```

```python
density2 =  np.array([[2,2,3,4,6,8],
                          [2,2,3,7,9,11],
                          [2,2,3,9,12,13],
                          [2,3,7,13,13,15],
                          [6,10,13,15,18,20],
                          [9,11,13,16,20,26]])
```

```python
initChoice2 = np.zeros(330)
l = 0
p = 0.5
for i in range(6):
    for j in range(6):
        density = density2[i,j]
        for k in range(density):
            initChoice2[l] = np.random.binomial(1, p*density/26,1)
            l += 1
```
